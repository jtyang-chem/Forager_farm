```python
import pyautogui
import time
import random
import pydirectinput as pdi
pyautogui.FAILSAFE = False
global pos_left
global pos_right
global pos_box
#global n_orb

pos_left = (639, 516)
pos_right = (958, 528)
#pos_box = (788, 429)
pos_box = (867, 622)
n_orb = 2440

def find_and_click(img):
    clickTimes =1
    #location=pyautogui.locateCenterOnScreen(img,confidence=0.9)
    location=pyautogui.locateCenterOnScreen(img)
    pyautogui.click(location.x,location.y,clicks=clickTimes,interval=0.2, duration=0.2)
    time.sleep(0.3)
    pyautogui.click(location.x,location.y,clicks=clickTimes,interval=0.2, duration=0.2)
    #pyautogui.click(location.x,location.y,clicks=clickTimes,interval=0.2,duration=0.2)
    
def test_find_and_click():
    img = "left.png"
    find_and_click(img)
    
def move_mouse_left():
    #x, y = (639, 516)
    x, y = pos_left
    pyautogui.moveTo(x, y)
    
def move_mouse_right():
    x, y = pos_right
    pyautogui.moveTo(x, y)
    
def add_hp():
    move_mouse_left()
    x, y= pyautogui.position()
    #clickTimes = 2
    #pyautogui.click(x,y,clicks=clickTimes,interval=0.3, duration=1)
    time.sleep(0.3)
    left_click(0.2)
    pyautogui.move(0, -20)
    time.sleep(2)
    #pdi.click(duration = 0.2)
    #print("one click")
    left_click(0.0)
    time.sleep(0.3)
    #pyautogui.click(x,y,clicks=clickTimes,interval=0.2, duration=0.2)
    #time.sleep(0.3)
    
def add_17hp():
    for i in range(17):
        add_hp()
        
def left_click(duration =0.8):
    pyautogui.mouseDown()
    #pdi.mouseDownwn()
    time.sleep(duration)
    pyautogui.click()

def exist_item(item=""):
    img = f"{item}.png"
    location=pyautogui.locateCenterOnScreen(img)
    if location == None:
        return False
    else:
        return True
    
def open_box():
    global pos_box
    x_box, y_box = pos_box
    pyautogui.moveTo(x_box, y_box)
    time.sleep(0.5)
    pyautogui.press('e')
    time.sleep(0.5)
    
def consume_hp():
    open_box()
    l, r = check_card()
    #print(l,r)
    choose_card(l,r)
    #pyautogui.press('e')
    
def select_card(pos):
    if pos == "left":
        move_mouse_left()
    else:
        move_mouse_right()
    left_click(0.0)
        
def choose_card(l,r):
    cards = [l, r]
    #card_set = ["attack", "termination", "potion"]
    for item in ["attack", "termination", "potion", "others"]:
        if item in cards:
            idx = cards.index(item)
            if idx == 0:
                select_card("left")
            else:
                select_card("right")
            break
    

def check_card():
    move_mouse_left()
    #time.sleep(0.5)
    l = "others"
    for item in ["attack", "termination", "potion"]:
        if exist_item(item):
            l = item
            break
    r = "others"
    move_mouse_right()
    time.sleep(0.5)
    for item in ["attack", "termination", "potion"]:
        if exist_item(item):
            r = item
            break
    return l,r

def consume_17hp():
    for i in range(17):
        consume_hp()

#time.sleep(5)
#print( pyautogui.size())
#print(pyautogui.position())
#test_find_and_click()
#move_mouse_left()
#move_mouse_right()
#pyautogui.press('e')
#for i in range(n_orb):
#    consume_hp()
#    add_hp()
    
#print(check_card())
def farm(n_orb):
    for i in range(n_orb):
        #consume_hp()
        add_hp()

def show_position():
    time.sleep(5)
    print(pyautogui.position())

#show_position()
farm(n_orb)
```
